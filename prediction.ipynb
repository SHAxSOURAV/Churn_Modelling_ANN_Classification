```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
##Load the ANN train model, scakler andd one hot model
model = load_model('model.h5')

##load the encoder and scaler
with open('labe_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)
with open('onehot_encoder_geo.pkl', 'rb') as file:
    label_encoder_geo = pickle.load(file)
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance':60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1, 
    'EstimatedSalary': 50000
}
```

```python
geo_encoded= label_encoder_geo.transform([[input_data["Geography"]]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df.head()
```

```python
df = pd.DataFrame([input_data])
```

```python
input_df = pd.concat([df.reset_index(drop=True),geo_encoded_df], axis =1)
input_df
```

```python
##Encode my catagorical Variable
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
input_df = input_df.drop('Geography',axis =1)
```

```python
input_df
```

```python
##Saclling the input data
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
#prediction 
prediction=model.predict(input_scaled)
prediction_proba = prediction[0][0]
```

```python
if prediction_proba > 0.5:
    print("The Customer will probably Churn")
else:
    print("The Customer will not likely to Churn")
```
