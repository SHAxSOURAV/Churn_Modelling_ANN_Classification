```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'], axis = 1)
data.head()
```

```python
labe_enco_gender = LabelEncoder()
data['Gender']=labe_enco_gender.fit_transform(data['Gender'])
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_enco= OneHotEncoder()
geo_encoded= one_hot_enco.fit_transform(data[["Geography"]])
geo_encoded
```

```python
one_hot_enco.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(geo_encoded.toarray(),columns=one_hot_enco.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df.head()
```

```python
data = pd.concat([data.drop("Geography", axis=1),geo_encoded_df], axis =1)
data.head()
```

```python
with open("labe_encoder_gender.pkl", 'wb') as file:
    pickle.dump(labe_enco_gender, file)
with open('onehot_encoder_geo.pkl', 'wb')as file:
    pickle.dump(one_hot_enco,file)
```

```python
data.head()
```

```python
X= data.drop("Exited", axis = 1)
Y = data['Exited']

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2,random_state=42)

scaler= StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('scaler.pkl', 'wb' ) as file:
    pickle.dump(scaler,file)
```

```python
##Trainig a anN
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
(x_train.shape[1],)
```

```python
x_train.shape[1]
```

```python
##Build our ANN Model
model=Sequential(
    [ 
        Dense(64, activation='relu', input_shape = (x_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1,activation='sigmoid')
    ]
)
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
##Compile the model
model.compile(optimizer=opt,loss='binary_crossentropy', metrics =['accuracy'])
```

```python
##Setup the Tensorboard
log_dir='logs/fit/' + datetime.datetime.now().strftime("%Y%M%D-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
##Setup Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
##Train the model
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs =100,callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
